```python
%%sql 

SELECT host_name, ROUND(avg(review_scores_rating)) as Average_Rating, ROUND(avg(availability_365)) as Average_Availability
  FROM Listings
 WHERE host_is_superhost = 't'
 GROUP BY host_id, host_name
HAVING avg(review_scores_rating) > 0
 ORDER BY avg(review_scores_rating) DESC, avg(availability_365) DESC
 LIMIT 5;
```

```python
import psycopg2 as pg2
con = pg2.connect(database='airbnb', user='isdb')
con.autocommit = True
cur = con.cursor()

import matplotlib.pyplot as plt
%matplotlib inline

query = '''
  SELECT zipcode, avg(price::numeric)
    FROM Listings
   WHERE zipcode::integer > 0
GROUP BY zipcode
HAVING count(id) > 10
ORDER BY avg(price::numeric) DESC
LIMIT 5
   '''

cur.execute(query)
rows = cur.fetchall()

zipcode = [zc for (zc, prc) in rows]
price = [prc for (zc, prc) in rows]
width = 0.35 

fig, ax = plt.subplots()
rects1 = ax.bar(zipcode, price, width, color='b')
ax.set_ylabel('Average price ($)')
ax.set_xlabel('SF Zipcode')
ax.set_title('Top 5 Zipcodes with Highest Average Price')


plt.show()
```

```python
import gmplot
dbcon = pg2.connect(database='airbnb', user='isdb')
dbcon.autocommit = True
cur = dbcon.cursor()

query = '''
  SELECT latitude, longitude
    FROM Listings
    GROUP BY latitude, longitude
   '''
cur.execute(query)
lat_long = cur.fetchall()

latitude = [lat for (lat, lon) in lat_long]
longitude = [lon for (lat, lon) in lat_long]

gmap = gmplot.GoogleMapPlotter(latitude[0],longitude[0], 12.5)
gmap.heatmap(latitude, longitude)
gmap.draw("heatplot.html")


```

```python
%%sql

SELECT latitude, longitude, avg(weekly_price::numeric)
  FROM Listings
 WHERE weekly_price::numeric > 0
 GROUP BY latitude, longitude;
         
```

```python
%%sql

SELECT latitude, longitude, avg(price::numeric)
  FROM Listings
  WHERE id in (SELECT id 
                FROM Listings
               WHERE number_of_reviews in (SELECT DISTINCT max(number_of_reviews)
                                             FROM Listings
                                             GROUP BY latitude, longitude, id
                                             HAVING avg(weekly_price::numeric)>0))
        and price::numeric > 0
  
 GROUP BY latitude, longitude

 

```

```python
%%sql 

SELECT neighbourhood
  FROM Listings
 GROUP BY neighbourhood
HAVING avg(review_scores_rating) > 0
 ORDER BY avg(review_scores_rating) DESC
 LIMIT 1;
```
